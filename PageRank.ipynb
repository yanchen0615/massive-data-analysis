```python
from pyspark import SparkContext
from pyspark import SparkConf
```

```python
conf=SparkConf().set("spark.default.parallelism", 4)
sc = SparkContext(conf=conf)
```

```python
def mapper1(line):
    wordlist = line.split("\t")
    key = wordlist[0]
    value = wordlist[1]
    maplist = []
    maplist.append((key, value))
    return maplist
```

"mapper1" is used to deal with the raw data.  

First, load the data, split it by "\t" and save as "wordlist".

Secondly, put the first value which represents the source and the second value which serves as the distination into the same parentheses.

```python
def reducer1(x,y):
    return x+y
```

'reducer1' is used to sum the elements which target to the same distination.

```python
def reducer2(x,y):
    return x+','+y
```

'reducer2' is used to seperate the distinations from the same source by ",".

```python
data = sc.textFile('p2p-Gnutella04.txt')
```

```python
data = data.flatMap(mapper1)
```

```python
datalist = data.reduceByKey(reducer2).collect()
```

```python
def mapper2(line):
    start = line[0]
    target = line[1]
    wordlist = target.split(',')
    maplist = []
    maplist.append((int(start),len(wordlist),wordlist))
    return maplist
```

"mapper2" transforms the raw data to the format that contains source, degree and distinations.

```python
for i in range(len(datalist)):
    temp = datalist.pop(0)
    datalist.extend(mapper2(temp))
```

"only_out" is a list of the nodes that have no in-degree.

```python
only_out = list(set(sourcelist).difference(set(distination)))
```

"unitlist" is a list that contains both in-degree nodes and out-degree nodes.

```python
unitlist = list(set(sourcelist).union(set(distination)))
```

```python
N = len(unitlist)
```

```python
beta = 0.8
```

Build a new list, called "rold".

```python
rold = []

for i in range(min(unitlist), N, 1):
    a = 1/N
    rold.append((i, a))
```

"rold" is filled with the initial value, 1/N.

```python
def mapper3(line, rold):
    maplist = []
    source = line[0]
    degree = line[1]
    dist = line[2]
    
    rold.sort(key = lambda x:x[0]!=source)
    value = rold[0][1]
    
    a = beta*value/degree
    for i in range(len(dist)):
        maplist.append((dist[i],a))
    return maplist
```

"mapper3" is a funciton that pass the value of the source that is multiplated by beta, beta*value/degree, to the distinations.

The command, "rold.sort(key = lambda x:x[0]!=source)", finds the list which contains the value of the source.

In addition, the command, "value = rold[0][1]", saves the value of the resource in "value".

```python
outlist = []
for j in range(len(datalist)):
    line = datalist[j]
    maplist = mapper3(line, rold)
    outlist.extend(maplist)
```

Transform each line of datalist to the format, (distination, passed value) and save in outlist.

```python
outlist
```

Sum up the values of each distination by the following steps.

Step 1: Transform the outlist from list type to rdd type.

```python
rdd1 = sc.parallelize(outlist)
```

Step 2:Sum up the values of each distination by the reducer, "reducer1".

```python
rnew = rdd1.reduceByKey(reducer1).collect()
```

```python
rnew
```

If the number of in-degree is 0, set the value of the node to 0.

```python
for i in range(len(only_out)):
    rnew.append((only_out[i],0))
```

Aggregate each numbers in "rnew" to re-insert the leaked pagerank.

```python
S = 0
for i in range(len(rnew)):
    S = S+rnew[i][1]
```

Add (1-S)/N to each terms in "rnew" and save them in a new list, called "r_new".

```python
r_new = []
for i in range(len(rnew)):
    a = int(rnew[i][0])
    b = rnew[i][1]+(1-S)/N
    r_new.append((a,b))
```

```python
r_new
```

Check whether the process is convergent by the value, called "diff" which is composed with each diffirence between "r_new" and "rold".

```python
diff = 0
for i in range(len(r_new)):
        diff = diff+r_new[i][1]-rold[i][1]
```

The term, r_new, plays the role of "rold" in the next itration.

```python
rold = r_new
```

```python
rold
```

Execute the previous computations until the process reaches the state of convergence.

```python
while diff>0.00001:
    outlist = []
    for j in range(len(datalist)):
        line = datalist[j]
        maplist = mapper3(line, rold)
        outlist.extend(maplist)
    
    rdd1 = sc.parallelize(outlist)
    rnew = rdd1.reduceByKey(reducer1).collect()
    S = 0
    for i in range(len(rnew)):
        S = S+rnew[i][1]
    r_new = []
    for i in range(len(rnew)):
        a = int(rnew[i][0])
        b = rnew[i][1]+(1-S)/N
        r_new.append((a,b))
    diff = 0
    for i in range(len(rold)):
        diff = diff+r_new[i][1]-rold[i][1]
    rold = r_new
```
