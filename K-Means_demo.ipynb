```python
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName('appName').setMaster('local')
sc = SparkContext(conf=conf)
```

Define Euclidean distance

```python
def Euclidean(line1, line2):
    #line1 = c1, line2 = data[i]
    distance = []
    for i in range(len(line1)):
        di = (float(line1[i])-float(line2[i]))**2
        distance.append(di)
    return sum(distance)**0.5
```

Define Manhattan distance

```python
def Manhattan(line1, line2):
    #line1 = c1, line2 = data[i]
    distance = []
    for i in range(len(line1)):
        di = abs(line1[i]-line2[i])
        distance.append(di)
    return sum(distance)
```

Part 1

Classification

```python
classification = []
for i in range(len(listdata)):
    dlist = []
    #calculate the distance between the point to each centers
    for j in range(len(list1)):
        d = Euclidean(listdata[i], list1[j])
        dlist.append(d)
    #choose the minimum
    classify = dlist.index(min(dlist))
    classification.append((classify, [(i, min(dlist))]))
```

Part 2

Recalibrate the centroids

```python
def reducer1(x,y):
    return x+y
```

```python
classification= sc.parallelize(classification).reduceByKey(reducer1).collect()
```

Caculate the new centroids

```python
def newcenter(line):
    center = []
    for m in range(dim):
        counter = 0
        for i in range(len(line)):
            counter = counter + float(listdata[i][m])
        center.append(counter/len(line))
    return center
```

```python
new_center = []
for i in range(len(classification)):
    line = classification[i][1]
    center = newcenter(line)
    new_center.append((center))
```

Calculate the cost

```python
cost = []
for i in range(len(classification)):
    #i represents the group
    for j in range(len(classification[i][1])):
        #j represents index
        cost.append(classification[i][1][j][1])
cost = sum(cost)        
```

```python
def high_dim_mean(index_list):   
    mean = []
    for d in range(dim):
        counter = 0
        for j in range(len(index_list)):
            counter+=float(listdata[index_list[j]][d])
            #print(counter)
        m = counter/dim
        mean.extend([m])
    return mean
```

```python
def new_center(line):
    newcenter = []
    index = []
    for i in range(len(line)):
        index.extend([line[i][0]])
    newcenter = high_dim_mean(index)
    return newcenter
```
