```python
import findspark
findspark.init()
import pyspark
```

```python
from pyspark import SparkContext
from pyspark import SparkConf
conf = SparkConf().set("spark.default.parallelism", 4)
sc = SparkContext(conf=conf)
```

```python
def get_set(string, k):
    set1 = []
    string = ''.join(string)
    list1 = string.split(' ')
    for i in range(len(list1)-k+1):
        set1.append(list1[i+0:i+k])
    return set1
```

# use 3-shingles

```python
k = 3
```

# try 001.txt

```python
str1 = sc.textFile('dataset/001.txt')
```

```python
set001 = str1.flatMap(lambda j: get_set(j, k)).collect()
```

# 002.txt~101.txt

```python
set002_101 = []
for i in range(2,102):
    seti = []
    if i<10:
        string_index = ''.join(['dataset/00',str(i),'.txt'])
    elif 10<=i<100:
        string_index = ''.join(['dataset/0',str(i),'.txt'])
    else:
        string_index = ''.join(['dataset/',str(i),'.txt'])
    file = sc.textFile(string_index)
    seti = file.flatMap(lambda j: get_set(j, k)).collect()
    set002_101.append(seti)
```

# Get general set

```python
general_set = set001
for i in range(len(set002_101)):
    general_set.extend(set002_101[i])
```
